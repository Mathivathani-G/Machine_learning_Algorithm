```python
import numpy as np
from sklearn.svm import SVR
from sklearn.datasets import load_diabetes
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt

# Load the diabetes dataset
diabetes = load_diabetes()

# Select the feature columns (columns 4 to 9) as X
X = diabetes.data[:, 3:8]

# Select the blood pressure column (column 3) as the target variable y
y = diabetes.target
#print(y)

# Create and train the SVR model
svr = SVR(kernel='rbf')
svr.fit(X, y)

# Make predictions
y_pred = svr.predict(X)

# Calculate mean absolute error and mean squared error
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)

# Plot the predicted values against the true values
plt.scatter(y, y_pred, color='darkorange')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='navy', lw=2, linestyle='--')
plt.xlabel('True Values')
plt.ylabel('Predictions')
plt.title('True vs Predicted Blood Pressure')
plt.show()
```
