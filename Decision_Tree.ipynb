```python
import pandas as pd 
import numpy as np
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,accuracy_score,classification_report,recall_score,f1_score,precision_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier 
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
balance_data = pd.read_csv('data4.csv',sep=',', header=None)
print("Dataset Length: ", len(balance_data))
print("Dataset Shape: ", balance_data.shape)
X = balance_data.values[:, 2:]
Y = balance_data.values[:, 1]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=100)
model = DecisionTreeClassifier(criterion="gini",random_state=100, max_depth=3, min_samples_leaf=5)
model.fit(X_train, y_train) 
y_pred = model.predict(X_test)
print("Predicted values:")
#TO CHECK THE PREDICTED VALUES:
"""for i in range(len(X_test)):
    print(X_test[i],"-",y_test[i])"""
print("Accuracy : ",accuracy_score(y_test, y_pred)*100)
precision = precision_score(y_test, y_pred, pos_label='M')
recall = recall_score(y_test, y_pred, pos_label='M')
f1 = f1_score(y_test, y_pred, pos_label='M')
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
cm = confusion_matrix(y_test, y_pred, labels=['B', 'M'])

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['B', 'M'])
disp.plot()
plt.title('Confusion Matrix')
plt.show()
```
